# cleaning the data

```python
#checking for null values

dataset.isnull().sum()
```

* there are no missing values

```python
dataset=dataset.drop(['Country', 'Postal Code'], axis=1)
```

```python
dataset1=dataset.rename(columns={'Sub-Category':'Sub_Category'})
```

# visualization

```python
#Relationship Analysis

corelation=dataset1.corr()
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot = True)
```

```python
dataset1['State'].unique()
```

```python
state_group=dataset1.groupby('State').sum()
state_group.reset_index(inplace=True)
print(state_group)
```

```python
sns.barplot(x=state_group['Sales'], y=state_group['State'], order=state_group.sort_values('Sales',ascending=False).State)
plt.title('State with highest sales', size=25)
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('State', size=20)
plt.ylabel('Sales', size=20)
plt.rcParams["figure.figsize"] = (18,20)
```

(i) Clifornia, New York and Texas top the states with most sales

(ii) Maine, West Virginia and North Dakota are the states with least sales

```python
shipmode_group=dataset1.groupby('Ship Mode').sum()
shipmode_group.reset_index(inplace=True)
print(shipmode_group)
```

```python
sns.histplot(x=dataset1['Ship Mode'])
plt.title('Most Preferred Ship Mode', size=15)
plt.rcParams["figure.figsize"] = (6,6)
```

(i) Standard class is the most preferred Ship Mode.

```python
shipmode_group.plot(x='Ship Mode', y=['Profit', 'Sales'], kind='bar')
plt.rcParams["figure.figsize"] = (4,3)
plt.legend(loc='upper left')
```

(i) Standard class ship mode gives the most sales and pofit.

(ii) Same Day ship mode gives the least sales as well as the least profit.

```python
region_group=dataset1.groupby('Region').sum()
region_group.reset_index(inplace=True)
print(region_group)

sns.barplot(x=region_group['Sales'], y=region_group['Region'], order=region_group.sort_values('Sales',ascending=False).Region)
plt.title('Region with highest sales', size=15)
```

(i) West region and East region are the two regions with the most sales

```python
sns.displot(dataset1['Quantity'], bins=14, kde=True)
plt.title('Quantity Distribution plot', fontsize=16)
plt.show()
```

(i) Highest quantities ordered are within the range of 2 to 4

##### Profit and Sales by Category

```python
category_group=dataset1.groupby('Category').sum()
category_group.reset_index(inplace=True)
print(category_group)
```

```python
category_group.plot(x='Category', y=['Profit', 'Sales'], kind='bar')
plt.rcParams["figure.figsize"] = (8,6)
plt.legend(loc='upper left')
```

(i) The most profits are coming from the Technology category.

(ii) Profit from Furniture category is very low compared to others.

##### Profit and Sales by Furniture sub-category

```python
furniture_category=dataset1[dataset1['Category']=='Furniture']

furniture_sub_category=furniture_category.groupby('Sub_Category').sum()
furniture_sub_category.reset_index(inplace=True)
print(furniture_sub_category)
```

```python
plt.rcParams["figure.figsize"] = (8,10)
furniture_sub_category.plot(x='Sub_Category', y=['Profit', 'Sales'], kind='bar')
plt.show()
```

(i) Both tables and Bookcases sub-category under Furniture are in losses.

##### Profit and Sales by sub-categories

```python
sub_category=dataset1.groupby(by=['Category','Sub_Category']).sum()
sub_category.reset_index(inplace=True)

sns.set(rc = {'figure.figsize':(12,8), 'axes.facecolor':'white'})
sns.barplot(x=sub_category['Profit'], y=sub_category['Sub_Category'], hue=sub_category['Category'], dodge=False, order=sub_category.sort_values('Profit',ascending=False).Sub_Category)
plt.title('Profits by Sub_Category', size=25)
plt.legend(loc='right')
```

(i) Tables and Bookcases under Furniture category are in losses.

(ii) Supplies sub-category under Office Supplies is also in negative profit.

(iii) Technology is the best category in terms of profits as the top three sub-categories are under Furniture category. 

##### Profit vs Discount Relationship

```python
discount_group=dataset1.groupby('Discount').sum()
discount_group.reset_index(inplace=True)
print(discount_group)
```

```python
plt.rcParams["figure.figsize"] = (13,7)
plt.stackplot(discount_group.Discount, discount_group.Profit, color='skyblue')
plt.title('Profit and Discount Relationship')
plt.xlabel('Discount')
plt.ylabel('Profit')
```

* discounts more than 30% result in losses

##### Discount and Sales by category

```python
print(category_group)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax2 = ax.twinx()
ax.set_title('Discount & Sales on different category items', size=16)
ax.set_xlabel('Category', size=14)
ax.plot(category_group['Category'], category_group['Sales'], color='green')
ax2.plot(category_group['Category'], category_group['Discount'], color='blue')
ax.set_ylabel('Sales', size=14)
ax2.set_ylabel('Discount', size=14)
ax.legend(['Sales'], bbox_to_anchor=(0.13,1))
ax2.legend(['Discount'], bbox_to_anchor=(0.13,0.91))
ax.yaxis.grid(color='lightgray', linestyle='dashed')
plt.tight_layout()
plt.show()
```

(i) Discount on Office Supplies category has no effect on the sales.

(ii) Technology has the lowest discount yet outperformed other categories in terms of sales.

##### Discount and Profit by category

```python
fig, ax = plt.subplots(figsize=(12,5))
ax2 = ax.twinx()
ax.set_title('Discount & Profit on different category items', size=16)
ax.set_xlabel('Category', size=14)
ax.plot(category_group['Category'], category_group['Profit'], color='green')
ax2.plot(category_group['Category'], category_group['Discount'], color='blue')
ax.set_ylabel('Profit', size=14)
ax2.set_ylabel('Discount', size=14)
ax.legend(['Profit'], bbox_to_anchor=(0.13,1))
ax2.legend(['Discount'], bbox_to_anchor=(0.13,0.91))
ax.yaxis.grid(color='lightgray', linestyle='dashed')
plt.tight_layout()
plt.show()
```

(i) Even though the discount on Technology is low, it has no effect on the profit as profit continues to rise.

# Summaries

* The best performing category in terms of sales and profit is the Technology category while the least is Furniture.
* California, New York and Texas are the top three states with most sales. North Dakota, West Virginia and Maine being the bottomed three states.
* The West region top the list of region with most sales.
* The highest ordered quantities are within the range of 2 to 4.
* The most preferred Ship Mode is Standard Class, the least preferred being Same Day.


#### Strengths
* The strength of this company is in Technology
* Copiers, Phones and Accessories products under the Technology category account for the most profits.
* Papers and Binders products are also yielding good profits.

#### Weaknesses
* Tables and Bookcases products are in great losses.
* Supplies products under the Office Supplies category is also resulting in losses.
* Discount on Office Supplies products are not leading to more sales.

## Recommendations

* Tables, Bookcases and Supplies products should be dropped or reconsidered.
* Discounts on products should not be more than 30% as anything higher than that results in losses.
* Discounts on Office Supplies items should be dropped.
* Policies surrounding Technology products should be maintained.
* Business policies around the states of California, New York and Texas should be maintaianed.
